# DataFrames

Los DataFrames son el caballo de batalla de  pandas y se inspiran directamente en el lenguaje de programación R. Podemos pensar en un DataFrame como un grupo de objetos Series juntos para compartir el mismo índice. ¡Usemos pandas para explorar este tema!

```python
import pandas as pd
import numpy as np
```

```python
from numpy.random import randn
np.random.seed(101)
```

```python
df = pd.DataFrame(randn(10,9),index='A B C D E F G H I J'.split(),columns='W X Y Z W X Y Z Z'.split())
```

```python
df
```

## Selección e indexación

Aprendamos los diversos métodos para obtener datos de un DataFrame

```python
df['W']
```

```python
# Pass a list of column names
df[['W','Z']]
```

```python
# Sintaxis SQL (NO RECOMENDADA!)
df.W
```


Las columnas de DataFrame son solo series

```python
type(df['W'])
```

### Creando una nueva columna:

```python
df['new'] = df['W'] + df['Y']
```

```python
df
```

### Eliminar columnas

```python
df.drop('new',axis=1)
```

```python
# ¡No en su lugar a menos que se especifique!
df
```

```python
df.drop('new',axis=1,inplace=True)
```

```python
df
```

También puede soltar filas de esta manera:

```python
df.drop('E',axis=0)
```

### Seleccionar filas

```python
df.loc['A']
```

O seleccione según la posición en lugar de la etiqueta

```python
df.iloc[2]
```

### Seleccionar subconjunto de filas y columnas

```python
df.loc['B','Y']
```

```python
df.loc[['A','B'],['W','Y']]
```

### Selección condicional

Una característica importante de los pandas es la selección condicional mediante notación de corchetes, muy similar a numpy:

```python
df
```

```python
df>0
```

```python
df[df>0]
```

```python
df[df['W']>0]
```

```python
df[df['W']>0]['Y']
```

```python
df[df['W']>0][['Y','X']]
```

Para dos condiciones puede utilizar | y & entre paréntesis:

```python
df[(df['W']>0) & (df['Y'] > 1)]
```

## Más detalles del índice

Analicemos algunas características más de la indexación, incluido el restablecimiento del índice o configurarlo de otra manera. ¡También hablaremos sobre la jerarquía de índices!

```python
df
```

```python
# Restablecer el índice predeterminado 0,1 ... n
df.reset_index()
```

```python
newind = 'CA NY WY OR CO'.split()
```

```python
df['States'] = newind
```

```python
df
```

```python
df.set_index('States')
```

```python
df
```

```python
df.set_index('States',inplace=True)
```

```python
df
```
